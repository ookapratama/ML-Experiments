## KS Statistic 

```python
def ks_statistic(y_true, y_proba):
    df_ks = pd.DataFrame({
        "y": y_true,
        "proba": y_proba
    }).sort_values("proba")

    cdf_good = np.cumsum(df_ks["y"] == 0) / (df_ks["y"] == 0).sum()
    cdf_bad = np.cumsum(df_ks["y"] == 1) / (df_ks["y"] == 1).sum()

    return np.max(np.abs(cdf_bad - cdf_good))

ks = ks_statistic(y_test.values, y_proba)
ks
```
